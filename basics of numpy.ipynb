```python
#boolean indexing
import numpy as np
a=np.arange(12).reshape(3,4)
print("a=\n",a)
b=a>4
print("b=\n",b)
a[b]=1
print("a=\n",a)
```

```python
#boolean indexing with strings
import numpy as np
names=np.array(['bob','joe','will','bob','will','joe','joe'])
data=np.arange(28).reshape(7,4)
print(names)
print("names bool= \n",names == 'bob')
print("names == 'Bob' \n",data[names == 'bob'])
print("names == 'Bob' \n",data[names == 'bob'])
print("names == 'Bob',3 \n",data[names == 'bob', 3])
```

```python
# select all except bob
import numpy as np
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.arange(28).reshape(7, 4)
print(data[names != 'Bob'])
print("\n",data[~(names == 'Bob')])
data[names != 'Bob']=-1
print("\ndata= \n",data)
```

```python
import numpy as pandas
arr=np.arange(32).reshape(8,4)
print("arr=\n",arr)
print("\narr[[1, 5, 7, 2]]= \n",arr[[1, 5, 7, 2]])
print("\narr[[1, 5, 7, 2], [0, 3, 1, 2]]= \n",arr[[1, 5, 7, 2], [0, 3, 1, 2]])
print("\narr[[1, 5, 7, 2]][:, [0, 3, 1, 2]]= \n",arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]])
print("\n arr[[-3, -5, -7]]\n",arr[[-3, -5, -7]])
arr[[1, 5, 7, 2]] = 1
print("\narr= \n",arr)
```
